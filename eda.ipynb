# EDA for the supplements dataset

This isn't called by the supplements application, but is useful to gain some insight about the supplements dataset. As it is relevant, it is being stored in this repo too.

```python
import os
import pandas as pd

dataset = r"/mnt/hdd/aiml339/data/supplements/data/tabular/products.csv"
```

```python
df = pd.read_csv(dataset)
```

```python
import matplotlib.pyplot as plt
```

```python
df.head()
```

```python
hist_data = df["company_name"].value_counts()

categories = hist_data.index
values = hist_data.values
```

```python
int(max(values)) # most represented company
```

```python
len(categories) # how many companies
```

```python
plt.bar(categories[0:25], values[0:25])
plt.xticks(rotation=90)
```
